```python
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq
from typing import TypedDict
```

```python
import getpass
import os

if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API key: ")
```

```python
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
class AgentState(TypedDict):
  query: str
  explanation: str
  summary: str
  code: str
  final_output: str
```

```python
from langchain_core.messages import HumanMessage

def explanation_agent(state: AgentState):
  query = state["query"]

  response = llm.invoke([
      HumanMessage(content=f"Explain this topic clearly: {query}")
  ])

  return {"explanation": response.content}
```

```python
def summary_agent(state: AgentState):
  query = state["query"]

  response = llm.invoke([
      HumanMessage(content=f"Give a short summary of {query}")
  ])

  return {"summary": response.content}
```

```python
def code_agent(state: AgentState):
  query = state["query"]

  response = llm.invoke([
      HumanMessage(content=f"Write a python code example for {query}")
  ])

  return {"code": response.content}
```

```python
def combine_agent(state: AgentState):

  final = f"""
  Explanation:
  {state['explanation']}

  Summary:
  {state['summary']}

  Code Example:
  {state['code']}
  """

  return {"final_output": final}
```

```python
Graph = StateGraph(AgentState)

Graph.add_node("explanation", explanation_agent)
Graph.add_node("summary", summary_agent)
Graph.add_node("code", code_agent)
Graph.add_node("combine", combine_agent)

Graph.add_edge(START, "explanation")
Graph.add_edge(START, "summary")
Graph.add_edge(START, "code")

Graph.add_edge("explanation", "combine")
Graph.add_edge("summary", "combine")
Graph.add_edge("code", "combine")

Graph.add_edge("combine", END)

Workflow = Graph.compile()
```

```python
result = Workflow.invoke({
    "query": "Convolutional Neural Networks"
})

print(result["final_output"])
```
