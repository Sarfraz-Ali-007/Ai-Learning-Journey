```python
import cv2
from google.colab.patches import cv2_imshow
```

**RESIZING IMAGE**

```python
image = cv2.imread('python_image.png')
```

```python
if image is None:
  print("Image not found")
else:
  print("Image loaded")

resized = cv2.resize(image, (300, 300))

print("====ORIGINAL IMAGE====")
cv2_imshow(image)
print("====RESIZED IMAGE====")
cv2_imshow(resized)

cv2.imwrite("resized_output.png", resized)
```

**CROPPING IMAGE**

```python
if image is not None:
  cropped = image[100:200, 50:150]

  print("====ORIGINAL IMAGE====")
  cv2_imshow(image)
  print("====CROPPED IMAGE====")
  cv2_imshow(cropped)
```
