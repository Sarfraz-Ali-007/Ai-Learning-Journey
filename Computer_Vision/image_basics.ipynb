```python
import cv2
from google.colab.patches import cv2_imshow
```

**LOAD IMAGE**

```python
#Load Image

image = cv2.imread("python_image.png")
```

```python
if image is None:
  print("Error: Image not found")

else:
  print("Image loaded successfully!")
```

**DISPLAY IMAGE**

```python
#Display Image

if image is not None:
  cv2_imshow(image)
else:
  print("Could not load the image")
```

**SAVE IMAGE**

```python
#Save Image

if image is not None:
  success = cv2.imwrite("output_python.png", image)
  if success:
    print("Image saved succesfully as 'output_python.png'")
  else:
    print("Failed to save an image")
else:
  print("Error: Could not load image")
```

**IMAGE DIMENSIONS**

```python
#Image Dimensions

if image is not None:
  h, w, c = image.shape
  print(f"Image Loaded:\nHeights: {h}\nWidth: {w}\nChannels: {c}")
else:
  print("Could not load image")
```

**GRAYSCALE CONVERSION**

```python
#GrayScale Conversion

if image is not None:
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  cv2_imshow(gray)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
else:
  print("Could not load the image")
```
